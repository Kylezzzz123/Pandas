```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('beijing_tianqi_2018.csv')
```

```python
df.loc[:, 'bWendu'] = df['bWendu'].str.replace('℃', '').astype('int32')
df.loc[:, 'yWendu'] = df['yWendu'].str.replace('℃', '').astype('int32')
```

```python
df.head()
```

```python
df['month'] = df['ymd'].str[:7]
df.head()
```

```python
data = df.groupby('month')['bWendu'].max()
data
```

```python
data.plot()
```

```python
df.head()
```

```python
group_data = df.groupby('month').agg({'bWendu': np.max, 'yWendu':np.min, 'aqi': np.mean})
group_data
```

```python
group_data.plot()
```

```python
tianqi_data = df.groupby('month').agg({'tianqi': np.max})
tianqi_data.columns=['Weather']
tianqi_data
```

```python
bWendu_mean_month = df.groupby('month')['bWendu'].mean()
bWendu_mean_month
```

```python
bWendu_mean_month.plot()
```
